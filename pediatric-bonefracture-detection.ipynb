## **Plotting sample images from each image folder**

```python
folders = {
    "images_part1": 5031,
    "images_part2": 5143,
    "images_part3": 4842,
    "images_part4": 5311
}

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

for ax, (folder, count) in zip(axes, folders.items()):
    image_list = os.listdir(folder)
    random_image = random.choice(image_list)
    image_path = os.path.join(folder, random_image)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    ax.imshow(image)
    ax.set_title(folder)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
